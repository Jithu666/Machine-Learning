Multiple Linear Regression

Import the Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load the Dataset

```python
df = pd.read_csv("D:/AIML/Machine Learning/Part 2 - Regression/Section 5 - Multiple Linear Regression/Python/50_Startups.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
df.head()
# print(X)
```

```python
print(y)
```

Encoding the Categorical values

```python
# State column has Categorical values it needs to be encoded into Numerical 
# values through OneHotEncoding

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

Spit the Dataset into Training and Test Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Training the Multiple Linear Regression Model on the Training DataSet

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predicting the Test Set Results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test),1)),1))
```
